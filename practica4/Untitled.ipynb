```python
import spacy
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from rake_nltk import Rake
import pytextrank
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```python
with open("4.txt", "r", encoding="utf-8") as f:
    contenido = f.read()
inicio = re.compile(r"\*\*\*.*START OF THE PROJECT GUTENBERG.*\*\*\*")
fin = re.compile(r"\*\*\*.*END OF THE PROJECT GUTENBERG.*\*\*\*")

indiceInicio = re.search(inicio,contenido)
indiceInicio = indiceInicio.span()
indiceInicio = indiceInicio[1]
indiceFin = re.search(fin,contenido)
indiceFin = indiceFin.span()
indiceFin = indiceFin[0]
contenido = contenido[indiceInicio:indiceFin]
print(contenido)
```

```python
contenido = contenido.lower().replace(".", "") 
tokenizado = nltk.word_tokenize(contenido)
print(tokenizado)
```

```python
stop_words = set(stopwords.words('english'))
```

```python
tags = nltk.pos_tag(tokenizado) 
nouns = [word for (word, tag) in tags if tag == "NN"] 
from sklearn.feature_extraction.text import TfidfVectorizer 
vectorizer = TfidfVectorizer() 
tfidf = vectorizer.fit_transform([contenido]) 

# Get the top 3 most important nouns 
top_nouns = sorted(vectorizer.vocabulary_, key=lambda x: tfidf[0, vectorizer.vocabulary_[x]], reverse=True)[:3] 

# Print the top 3 keywords 
print(top_nouns)
```

```python
def leerDocumentos(numDocumentos):
    corpus = []
    for i in range(numDocumentos):
        doc = str(i+1) + ".txt"
        with open(doc, "r", encoding="utf-8") as f:
            contenido = f.read()
        indiceInicio = re.search(inicio,contenido)
        indiceInicio = indiceInicio.span()
        indiceInicio = indiceInicio[1]
        indiceFin = re.search(fin,contenido)
        indiceFin = indiceFin.span()
        indiceFin = indiceFin[0]
        contenido = contenido[indiceInicio:indiceFin]
        corpus.append(contenido)
    return corpus
    
```

```python
corpus = leerDocumentos(6)
```

```python
def limpiarTexto(corpus):
    tokensSinSS = []
    for doc in corpus:
        doc = doc.lower().replace(".", "") 
        tokens = word_tokenize(doc)
        texto_sin_ss = [w for w in tokens if not w.lower() in stop_words]
        texto_con_tags = nltk.pos_tag(texto_sin_ss)
        texto_con_tags2 = [word for (word, tag) in texto_con_tags if tag == "NN"] 
        
        #StrA = " ".join(texto_con_tags2)
        tokensSinSS.append(texto_con_tags2)
    return tokensSinSS
    
```

```python
nuevoCorpus = limpiarTexto(corpus)
```

```python
print(nuevoCorpus[0])
```

```python
#Funcion para extraer keywords usando Rake
def extraerConRake(corpus):
    extraccion = Rake()
    keywordsTotal = []
    for doc in corpus:
        extraccion.extract_keywords_from_text(doc)
        keywords = extraccion.get_ranked_phrases_with_scores()
        keywordsTotal.append(keywords[0:5])
    return keywordsTotal

keywordsRake = extraerConRake(corpus)
```

```python
for kw in keywordsRake:
    print(kw)
```

```python
#funcion para obtener keywords con textrank
def extraerConTextRank(corpus):
    nlp = spacy.load("en_core_web_sm")
    nlp.add_pipe("textrank")
    frases = []
    # examine the top-ranked phrases in the document
    for doc in corpus:
        doc = nlp(doc)
        xdd= []
        cont = 0;
        for phrase in doc._.phrases:
            if cont > 4:
                break
            lol = phrase.rank, phrase.text
            xdd.append(lol)
                #print(phrase.text)
                #print(phrase.rank)
            cont = cont+1
        frases.append(xdd)
    return frases
kwTextRank = extraerConTextRank(corpus)
```
